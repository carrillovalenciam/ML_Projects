```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = cv2.imread('bodies.jpg')

#convert to grayscale

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_image, cmap='gray')
```

```python
def convertToRGB(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
haar_cascade_body = cv2.CascadeClassifier()
haar_cascade_body.load('./haarcascades/haarcascade_fullbody.xml')
```

```python
bodies_rects = haar_cascade_body.detectMultiScale(gray_image, scaleFactor = 1.3, minNeighbors=1)
print('Bodies found: ', len(bodies_rects))
for (x,y,w,h) in bodies_rects:
    print(x,y,w,y)
    cv2.rectangle(image, (x,y), (x+w,y+h), (0,255,0),2)
```
