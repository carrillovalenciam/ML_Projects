```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = cv2.imread('baby_image.jpg')

#convert to grayscale

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_image, cmap='gray')

```

```python
def convertToRGB(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
haar_cascade_face = cv2.CascadeClassifier()
```

```python
haar_cascade_face.load('./haarcascades/haarcascade_frontalface_default.xml')
```

```python
faces_rects = haar_cascade_face.detectMultiScale(gray_image, scaleFactor = 1.1, minNeighbors = 1)
```

```python
print('Faces found: ', len(faces_rects))
```

```python
for (x,y,w,h) in faces_rects:
    print(x,y,w,y)
    cv2.rectangle(image, (x,y), (x+w,y+h), (0,255,0),2)
```

```python
plt.imshow(convertToRGB(image))
```
